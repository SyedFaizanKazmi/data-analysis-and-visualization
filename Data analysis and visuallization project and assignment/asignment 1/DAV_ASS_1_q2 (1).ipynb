```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("House Pricing.csv")
df.head()
```

```python
bins = [0, 25, 46,df['age'].max()]
labels = ['young', 'middle-aged', 'old']
df['age_order'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
 df
```

```python
df['is_charles_river'] = np.where(df['chas'] >= 1, 1, 0)
```

```python
df
```

```python
col = df.select_dtypes(include=['number']).columns
 
for column in col:
    plt.figure(figsize=(9, 7))
    plt.boxplot(df[column], vert=True)
    plt.title(f'Box Plot for {column}')
    plt.xlabel('Value')
    plt.ylabel(column)
    plt.show()
```

```python
info = []
col = df.select_dtypes(include=['number']).columns
for column in col:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR 
    
    
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    info.append({
        'Column': column,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
      
    })
    
    plt.figure(figsize=(9, 7))
    plt.boxplot(df[column], vert=True)
    plt.title(f'Box Plot for {column}')
    plt.show()
 
```

```python
outlier_counts = {}
for column in df.columns:
    if df[column].dtype in [int, float]:
        outliers = df[column][(df[column] < lower_bound) | (df[column] > upper_bound)]
        count = len(outliers)
        outlier_counts[column] = count
        print(f'Column: {column}, Count: {count}')
```

```python
for info in info:
    print(f"Column: {info['Column']}, IQR: {info['IQR']}, Lower Bound: {info['Lower Bound']}, Upper Bound: {info['Upper Bound']} ")
```

```python
from scipy import stats
 
z_scores=stats.zscore(df[['crim','zn','indus','nox','rm','age','dis','rad','tax','ptratio','b','lstat','medv']])
threshold=3
data_cleaned=df[(z_scores < threshold).all(axis=1)]
```

```python
alpha=0.5
df['rm_smoothed'] = df['rm'].ewm(alpha=alpha ,adjust=False).mean()
df
```

```python
df['tax_normalized'] = (df['tax'] - df['tax'].min()) / (df['tax'].max() - df['tax'].min())
df['lstat_normalized'] = (df['lstat'] - df['lstat'].min()) / (df['lstat'].max() - df['lstat'].min())
```

```python
 df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

model = LinearRegression()
X = df[['rm']] 
y = df['medv'] 
model.fit(X,y)

y_predicted = (model.coef_*X) + model.intercept_
 
plt.scatter(X, y, label='Data Points')
plt.plot(X,y_predicted,color='red')
plt.show()
print('slope : ',model.coef_)
print('intercept : ',model.intercept_)
mse = mean_squared_error(y, y_predicted)
print("Mean squared error:", mse)
r2 = r2_score(y,y_predicted)
print("R-squared:", r2)
```

```python
print(f"Regression Equation: medv = {model.intercept_} + {model.coef_} * rm")
```

```python
#The regression equation indicates that as the number of rooms per dwelling increases, 
#the median home value also increases. This implies a positive linear relationship between these two variables.
# The slope of the equation represents the amount by which the house value increases when the number of rooms increases.
```
